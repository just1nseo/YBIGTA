#### quailty가 7 이상인 데이터 추출 (20개)

```python
data_sample[data_sample.quality >= 7].head(20)
```

#### 'strength' 컬럼추가 : alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (상위 30개만 출력)

```python
data['strength'] = data['alcohol'].apply(lambda x: 'strong' if x > 10 else 'weak')
data.head(30)
```

#### merge 실습을 위한 데이터 쪼개기


```python
df_x = data[['quality', 'fixed acidity', 'citric acid']].head(30)
```

```python
df_y = data[['quality', 'volatile acidity', 'chlorides', 'alcohol']].head(30)
```

```python
df_y = df_y[df_y.quality >= 7]
```

```python
len(df_y)
```

#### df_x와 df_y를 inner join시키기 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how='inner', on='quality')
```

#### df_x에서 df_y를 left join시키기 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how='left', on='quality')
```

#### NaN, 데이터 타입, 개수 확인

#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
pd.to_datetime(data1['birthDate'])
```

```python
data1['birthDate'] = pd.to_datetime(data1['birthDate'])
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
import datetime as dt
x = dt.datetime.now()

data1['age'] = data1['birthDate'].apply(lambda a: x - a)
data1.head()
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
data1['age'].fillna(data1['age'].mean(),inplace=True)
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
data1['heightCm'] = data1['height'].apply(lambda x : float(x.split('-')[0])*30.48 + float(x.split('-')[1])*2.54)
data1.head()
```
