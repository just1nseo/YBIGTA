### 2) 결측치 확인

```python
# Q2
null = df.isnull().sum()[df.isnull().sum() > 0].sort_values(ascending=False)
df_null = pd.DataFrame({"missing_num": null, "missing_percentage": (null / len(df) * 100).round(0)})
df_null = df_null.iloc[:18]
df_null
```

```python
# Q3
plt.figure(figsize = (15,4))
plot = sns.barplot(x = df_null.index, y = 'missing_percentage', data = df_null)
plot.set_xticklabels(plot.get_xticklabels(), rotation=45)
plt.show()
```
